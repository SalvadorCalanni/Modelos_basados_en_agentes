```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import funciones_MBA as fMBA
```

```python
#Par√°metros
n_neu = 50
pasos = 750
peso_inicial = 10
```

```python
largo_estim = 4
listas_pesos_un_estim4 = []

nucleo4, coclea = fMBA.iniciar_red (n_neu, peso_inicial)
pos = nx.circular_layout(nucleo4)
nx.draw(nucleo4, pos, with_labels=True)
plt.show()


nro_vertices_un_estim4 = []
nucleo4, coclea = fMBA.iniciar_red (n_neu, peso_inicial)
for i in range(pasos):

    estim = fMBA.generar_estimulo(coclea, largo_estim)

    edges_estimulo = fMBA.vertices_estimulo(nucleo4, estim)

    fMBA.actualizar_pesos(nucleo4, edges_estimulo, estim)

    fMBA.eliminar_pesos_negativos(nucleo4)

    nro_vertices_un_estim4.append(len(list(nucleo4.edges())))

pos = nx.circular_layout(nucleo4)
nx.draw(nucleo4, pos, with_labels=True)
plt.show()


listas_pesos_un_estim4.append([nucleo4[u][v]['weight'] for u, v in nucleo4.edges()])
```
